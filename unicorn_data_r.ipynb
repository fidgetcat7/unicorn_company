```r
# Read a CSV file named "data.csv"
unicorn_data <- read.csv("unicorn_companies_clean.csv")

# Print the first few rows of the data frame
head(unicorn_data)
```

```r
library(tidyverse)
```

```r
# Assuming your data is in a data frame called "unicorn_data"
# with columns "company_name" and "investors"

# Split the "investors" column into a list of investors
unicorn_data <- unicorn_data %>%
  mutate(investors = strsplit(Select_Investors, ","))

# Unnest the list of investors into individual rows
unicorn_data <- unnest(unicorn_data, investors)

#
# Count the number of times each investor appears
investor_counts <- unicorn_data %>%
  mutate(investors = tolower(investors)) %>%
  count(investors, name = "num_unicorns")

# Sort the results by the number of unicorns funded
investor_counts <- investor_counts %>%
  arrange(desc(num_unicorns))

# Print the top investors
View(unicorn_data)
```

```r
# Print the top investors
head(investor_counts,10)
```

```r
write.csv(unicorn_data, "unicorn_data.csv", row.names = FALSE)
```
